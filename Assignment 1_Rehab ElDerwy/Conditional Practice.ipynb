```python
# Prompts the user for a grade () and prints a performance rating based on predefined thresholds.

# Accept user input for grade
grade = float(input("Enter your grade (0-100): "))
    
# Determine performance rating based on grade
if grade >= 90:
    print("Performance: Excellent")
elif grade >= 75:
    print("Performance: Good")
elif grade >= 50:
    print("Performance: Pass")
else:
    print("Performance: Fail")
```
